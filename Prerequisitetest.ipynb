Multiplications: nmk,
Additions: nm(k-1),
Explaination: As during matrix multiplication the row of former matrix moves from 1 to n and the column of later matrix moves from 1 to m. As during each such nm times, there are k multiplications and k-1 additions, total multiplications become nmk and total additions become nm(k-1).

```python
import numpy as np
import time

A = [[2,3,4]]
B = [[2],[3],[4]]

def mat_mul(A, B):
    n = len(A)
    k = len(B)
    m = len(B[0])
    C = [[0] * m for i in range(n)]
    for i in range(n):
        for j in range(m):
            for l in range(k):
                C[i][j] += A[i][l] * B[l][j]
    print(C)
start_brute = time.time()
mat_mul(A,B)
end_brute = time.time()
print(end_brute - start_brute)
start_numpy = time.time()
np.dot(A,B)
end_numpy = time.time()
print(end_numpy - start_numpy)
```

Partial differentiation wrt x = 2xy + (y^3)cos(x),
Partial differentiation wrt y = x^2 + 3(y^2)sin(x)

```python
import jax
import jax.numpy as jnp

def f(x, y):
    return x**2 * y + y**3 * jnp.sin(x)

def grads(x0,y0):
    grad_f_x = jax.grad(lambda x: f(x, y0))(x0)
    grad_f_y = jax.grad(lambda y: f(x0, y))(y0)

    return grad_f_x, grad_f_y

print(grads(3.0,4.0))
```

```python
class Student:
    def __init__(self, roll_no, name, marks):
        self.roll_no = roll_no
        self.name = name
        self.marks = marks
    def __repr__(self):
        return f"(Roll Number: {self.roll_no}, Name: {self.name}, Marks: {self.marks})"

class Branch:
    def __init__(self, branch):
        self.branch = branch
        self.students = []
    def student(self, student):
        self.students.append(student)
    def __repr__(self):
        return f"({self.branch},{self.students})"

class Year:
    def __init__(self, year):
        self.year = year
        self.branches = {}
    def branch(self, branch):
        self.branches[branch] = Branch(branch)
    def obtain_branch(self, branch):
        return self.branches.get(branch)
    def __repr__(self):
        return f"({self.year}, {self.branches})"

class Database:
    def __init__(self):
        self.years = {}
    def year(self, year):
        self.years[year] = Year(year)
    def obtain_year(self, year):
        return self.years.get(year)
    def __repr__(self):
        return f"({self.years})"

db = Database()
years = [2022, 2023, 2024, 2025]
for year in years:
    db.year(year)

year_2022 = db.obtain_year(2022)
year_2022.branch("Branch 1")
year_2022.branch("Branch 2")

branch1_2022 = year_2022.obtain_branch("Branch 1")
branch1_2022.student(Student(1, "A", {"Maths": 100, "English": 70}))
branch1_2022.student(Student(2, "B", {"Maths": 70, "English": 100}))

branch2_2022 = year_2022.obtain_branch("Branch 2")
branch2_2022.student(Student(3, "C", {"Maths": 70, "English": 70}))

print(db)
```

```python
import pandas as pd

matrix = np.random.uniform(1, 2, size=(20, 5))
df = pd.DataFrame(matrix, columns=['a', 'b', 'c', 'd', 'e'])

std = df.std()
max = std.max()
maxcol = std[std == max].index[0]
print(maxcol)

mean = df.mean(axis=1)
min = mean.min()
minrow = mean[mean == min].index[0]
print(minrow)
```

```python
def argmin(arr):
    min = 0
    for i in range(1, len(arr)):
        if arr[i] < arr[min]:
            min= i
    return min
arr = np.array([4,1,3])
print(argmin(arr))
print(np.argmin(arr))
```
